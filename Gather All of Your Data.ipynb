## Adding Markers

```python
# add markers to map
map_sby = folium.Map(location=[latitude, longitude], zoom_start=12, tiles='cartodbpositron')
for lat, lng, borough, neighborhood in zip(data_all['Latitude'], data_all['Longitude'], data_all['Borough'], data_all['District']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='green',
        
        fill_color='#3186cc',
        parse_html=False).add_to(map_sby)  
    
map_sby
```
